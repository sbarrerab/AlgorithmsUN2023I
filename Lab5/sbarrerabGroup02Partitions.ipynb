###Write a code that that produces  list of all partitions of the set of $[n]=\{ 1,2,3,4\cdots , n-1, n\}$ with $0 \le n \le 65$ and also list all the partitions for which the sum of elements of the partition is the same constant for all.

```python
def bell(n):
    # Función para generar los números de Bell usando programación dinámica
    bell_numbers = [[0] * (n + 1) for _ in range(n + 1)]
    bell_numbers[0][0] = 1

    for i in range(1, n + 1):
        bell_numbers[i][0] = bell_numbers[i - 1][i - 1]
        for j in range(1, i + 1):
            bell_numbers[i][j] = bell_numbers[i - 1][j - 1] + bell_numbers[i][j - 1]

    return bell_numbers[n][0]

def generate_partitions(n):
    # Función para generar todas las particiones de [n]
    partitions = []

    def generate_partition(remaining, partition):
        if remaining == 0:
            partitions.append(partition)
            return
        if not partition:
            start = 1
        else:
            start = partition[-1] + 1

        for i in range(start, n + 1):
            if remaining - i >= 0:
                generate_partition(remaining - i, partition + [i])

    generate_partition(n, [])
    return partitions

def constant_sum_partitions(n):
    # Función para enumerar particiones con suma constante
    partitions = generate_partitions(n)
    constant_sum_partitions = []

    for partition in partitions:
        if len(set(partition)) == 1:
            constant_sum_partitions.append(partition)

    return constant_sum_partitions

n = 65  # Puedes cambiar este valor según tus necesidades
bell_number = bell(n)
all_partitions = generate_partitions(n)
constant_sum_partitions_list = constant_sum_partitions(n)

print(f"Número de Bell para n={n}: {bell_number}")
print(f"Todas las particiones de [n]: {all_partitions}")
print(f"Particiones con suma constante: {constant_sum_partitions_list}")
```

# Write a code that produces the partitions with all the parts having same sum value, identify the formula for $n < 16$.

```python
def constant_partition(n):
    if n == 0:
        return [[]]
    if n == 1:
        return [[1]]

    partitions = []
    for k in range(1, n + 1):
        sub_partitions = constant_partition(n - k)
        for sub_partition in sub_partitions:
            if not sub_partition or sub_partition[0] >= k:
                partitions.append([k] + sub_partition)

    return partitions

n = 15  # Puedes cambiar este valor según tus necesidades
constant_partitions_list = constant_partition(n)

print(f"Particiones con suma constante para n={n}:")
for partition in constant_partitions_list:
    print(partition)
```
